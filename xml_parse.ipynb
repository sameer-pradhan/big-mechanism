```python
import xml.etree.ElementTree as ET
import os
import glob
import csv
import pandas as pd

filenames = glob.glob("data/*.nxml")
protein = {}

for filename in filenames:
    
    with open(filename, "r") as content:
        tree = ET.parse(content)
        root = tree.getroot()
        
        for span in root.findall('.//span[@type="participant"]'):
            #if key exists add value to existing list else make new key
            if span.get('intactdbid') in protein:
                protein[span.get('intactdbid')].append(span.text.encode('utf8'))
            else:
                protein[span.get('intactdbid')] = [span.text.encode('utf8')]
        
        for span in root.findall('.//span[@type="value"]' ):
            if span.get('intactdbid') in protein:
                protein[span.get('intactdbid')].insert(8, span.text.encode('utf8'))

        for span in root.findall('.//span[@type="unit"]'):
            if span.get('intactdbid') in protein:
                protein[span.get('intactdbid')].append(span.text.encode('utf8'))
                
        #print(root.tag), filename
        
#print protein

fields = [''] * 49
fields.insert(0,'interact id')


#returns a sorted list based on length of the value
def sort_dict(dict_):
    return [ ([k] + dict_[k]) for _, 
            k in sorted((len(v), k) for (k, v) in dict_.items())]


with open('interact.csv', 'w') as csvfile:
    
    #writes the columns
    writer = csv.writer(csvfile)
    writer.writerow(fields)
    
    protein_sort_lst = sort_dict(protein)
    
    #writes the rows
    for elmt in protein_sort_lst:    
        writer.writerow(elmt)
    
data = pd.read_csv('interact.csv')
data
    



```
